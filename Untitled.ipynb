```python
def get_location(file_path, line_number):
    line = linecache.getline(file_path, line_number).strip()
    location = [float(x) for x in line.split()]
    center = (int(location[0]), int(location[1]))
    return center

def get_santingwuyan_ratio(landmarks_dict, pic_name):
    l = landmarks_dict[pic_name]
    left_eye_left = l[1][0] - l[0][0]
    left_eye = l[2][0] - l[1][0]
    dist_eye = l[3][0] - l[2][0]
    right_eye = l[4][0] - l[3][0]
    right_eye_right = l[5][0] - l[4][0]
    width = l[5][0] - l[0][0]
    wuyan = [left_eye_left, left_eye, dist_eye, right_eye, right_eye_right]
    wuyan_ratio = [x/width for x in wuyan]

    eyebrow = (l[7][1] + l[8][1] + l[9][1] + l[10][1])/4
    upper = eyebrow - l[6][1]
    mid = l[11][1] - eyebrow
    lower = l[12][1] - l[11][1]
    length = l[12][1] - l[6][1]
    santing = [upper, mid, lower]
    santing_ratio = [x/length for x in santing]
    return wuyan_ratio, santing_ratio

landmarks_list = [6, 47, 43, 55, 51, 18, 1, 28, 32, 38, 42, 67, 12]
```

## 将 micro face analysis 的结果向量化

```python
def mapping_eye_shape(eye_shape, confidence, weight=5, mapping_dict={'ba01':0, 'ba02':-1, 'ba03':2, 'ba04':0, 'ba05':-2}):
    return weight*mapping_dict[eye_shape]*confidence

def mapping_eye_dist(eye_distance, confidence, weight=5, mapping_dict={'ad01':1, 'ad02':-1, 'ad03':0}):
    return weight*mapping_dict[eye_distance]*confidence

def mapping_nose(nose_shape, confidence, weight=5, mapping_dict={'ca01':-1, 'ca02':0, 'ca03':1, 'ca04':2}):
    return weight*mapping_dict[nose_shape]*confidence

def mapping_mouth_thickness(mouth, confidence, weight=3, mapping_dict={'da01':0, 'da02':1, 'da03':2, 'da04':-1, 'da05':-2}):
    return weight*mapping_dict[mouth]*confidence

def mapping_cheek_shape(cheek, confidence, weight=5, mapping_dict={'ea01':1, 'ea02':-1}):
    return weight*mapping_dict[cheek]*confidence

def mapping_facetype(circ, elliptic, long, prism, sq, triangle, weight=20):
    templist = [circ['value'], elliptic['value'], long['value'], prism['value'], sq['value'], triangle['value']]
    return [float(x) * weight for x in templist]

def mapping_lip_peak(lip_peak, confidence, weight=1, mapping_dict={'db01':0, 'db02':2, 'db03':1}):
    return weight*mapping_dict[lip_peak]*confidence

def mapping_face_vec(face_json):
    vec = [0, 0, 0, 0, 0]
    vec[0] = mapping_eye_shape(face_json['eye_shape']['value'], face_json['eye_shape']['confidence'])
    vec[1] = mapping_eye_dist(face_json['eye_distance']['value'], face_json['eye_distance']['confidence'])
    vec[2] = mapping_nose(face_json['nose_shape']['value'], face_json['nose_shape']['confidence'])
    vec[3] = mapping_mouth_thickness(face_json['mouth_thickness']['value'], face_json['mouth_thickness']['confidence'])
    vec[4] = mapping_cheek_shape(face_json['cheek_shape']['value'], face_json['cheek_shape']['confidence'])
    vec.extend(mapping_facetype(face_json['facetype_circular'], face_json['facetype_elliptic'], face_json['facetype_long'], face_json['facetype_prism'], face_json['facetype_square'], face_json['facetype_triangle']))
    return vec

def beauty_ugly_match(ugly_json, face_array, pic_name_list, top_k=3):
    json_obj = json.loads(ugly_json)
    face_json = json_obj['media_info_list'][0]['media_extra']['faces'][0]['face_attributes']
    ugly_vec = np.array(mapping_face_vec(face_json))
    norm = []
    for i in range(face_array.shape[0]):
        vec = face_array[i,:]
        norm.append(np.linalg.norm(vec-ugly_vec))
    norm = np.array(norm)
    sorted_index = np.argsort(norm)
    most_similar = sorted_index[:top_k]
    return [pic_name_list[x] for x in most_similar]
```

```python
def get_landmarks(pic_name):
    fa_marker = fa.FaceAlignment(fa.LandmarksType._2D, flip_input=False, device='cuda')
    input1 = io.imread('./Images/'+ pic_name)
    input1 = cv2.resize(input1, (350, 350))
    preds = fa_marker.get_landmarks(input1)
    return preds[0]

def get_face_analysis(landmarks, face_json):
    left_eye_brow_top = (landmarks[18][1] + landmarks[19][1] + landmarks[20][1]) / 3
    right_eye_brow_top = (landmarks[23][1] + landmarks[24][1] + landmarks[25][1]) / 3
    eye_brow_upperbound = (left_eye_brow_top + right_eye_brow_top) / 2
    face_length = landmarks[8][1] - eye_brow_upperbound
    face_width = landmarks[16][0] - landmarks[0][0]
    left_eye_length = landmarks[39][0] - landmarks[36][0]
    left_eye_width = (landmarks[41][1] + landmarks[40][1] - landmarks[37][1] - landmarks[38][1])*1.1/2
    right_eye_length = landmarks[45][0] - landmarks[42][0]
    right_eye_width = (landmarks[47][1] + landmarks[46][1] - landmarks[46][1] - landmarks[44][1])*1.1/2
    eye_distance = landmarks[42][0] - landmarks[39][0]
    eye_shape = mapping_eye_shape(face_json['eye_shape']['value'], face_json['eye_shape']['confidence'])
    nose_length = landmarks[33][1] - landmarks[27][1]
    nose_width = (landmarks[35][0] - landmarks[31][0]) * 1.5
    nose_shape = mapping_nose(face_json['nose_shape']['value'], face_json['nose_shape']['confidence'])
    mouth_thickness = mapping_mouth_thickness(face_json['mouth_thickness']['value'], face_json['mouth_thickness']['confidence'])
    mouth_width = landmarks[54][0] - landmarks[48][0]
    nose2lips = landmarks[51][1] - landmarks[33][1]
    lips2jaw = landmarks[8][1] - landmarks[57][1]
    lips_peak = mapping_lip_peak(face_json['lip_peak']['value'], face_json['lip_peak']['confidence'])
    cheek_shape = mapping_cheek_shape(face_json['cheek_shape']['value'], face_json['cheek_shape']['confidence'])
    mid_court = landmarks[33][1] - eye_brow_upperbound
    lower_court = landmarks[8][1] - landmarks[33][1]
    cheek_width = landmarks[12][0] - landmarks[4][0]
    mid_court_ratio = mid_court / face_length * 100
    lower_court_ratio = lower_court / face_length * 100
    eye_width_ratio = (left_eye_width + right_eye_width) / face_length * 50
    eye_length_ratio = (left_eye_length + right_eye_length) / face_width * 50
    eye_distance_ratio = eye_distance / face_width * 100
    nose_length_ratio = nose_length / face_length * 100
    nose_width_ratio = nose_width / face_width * 100
    nose2lips_ratio = nose2lips / face_length * 100
    lips2jaw_ratio = lips2jaw / face_length * 100
    mouth_width_ratio = mouth_width / cheek_width * 100
    cheek_ratio = cheek_width / face_width * 100
    risorius = float(face_json['risorius_yes']['value'])
    temple = float(face_json['temple_full']['value'])
    eye_arr = np.array([eye_shape, eye_width_ratio, eye_length_ratio, eye_distance_ratio])
    nose_arr = np.array([nose_shape, nose_width_ratio, nose_length_ratio])
    lips_arr = np.array([mouth_thickness, nose2lips_ratio, lips2jaw_ratio, mouth_width_ratio])
    facetype_arr = np.array([cheek_ratio, mid_court_ratio, lower_court_ratio, cheek_shape, risorius, temple])
    return eye_arr, nose_arr, lips_arr, facetype_arr

def eye_comparison(b, u):
    response = ['眼部分析结果：']
    diff = u - b

    if diff[0] < 0:
        response.append('眼型分析：你的眼睛太小，需要化大')
    else:
        response.append('眼型分析：你的眼睛已经够大了')

    response.append(f'眼宽分析：你的眼睛上下宽度得分{u[1]:.2f}, 美女得分{b[1]:.2f}')
    if diff[1] < 0:
        response.append(f'眼宽分析：你的眼睛上下宽度需要提升')
    else:
        response.append('眼宽分析：你的眼好宽哦')

    response.append(f'眼长分析：你的眼睛长度得分{u[2]:.2f}, 美女得分{b[2]:.2f}')
    if diff[2] < 0:
        response.append(f'眼长分析：你的眼睛长度需要提升，建议用眼线笔化长')
    else:
        response.append('眼长分析：你的眼好长哦')

    response.append(f'眼间距分析：你的眼间距得分{u[3]:.2f}, 美女得分{b[3]:.2f}')
    if diff[3] > 0:
        response.append(f'眼间距分析：你的眼间距需要缩短，建议修饰内眼角')
    else:
        response.append('眼间距分析：你的眼间距好窄哦')
    return response, diff
        

#def nose_comparison(b_arr, u_arr):

def drawline(img,pt1,pt2,color,thickness=1,style='dotted',gap=0):
    dist =((pt1[0]-pt2[0])**2+(pt1[1]-pt2[1])**2)**.5
    pts= []
    for i in np.arange(0,dist,gap):
        r=i/dist
        x=int((pt1[0]*(1-r)+pt2[0]*r)+.5)
        y=int((pt1[1]*(1-r)+pt2[1]*r)+.5)
        p = (x,y)
        pts.append(p)

    if style=='dotted':
        for p in pts:
            img = cv2.circle(img,p,thickness,color,-1)
    else:
        s=pts[0]
        e=pts[0]
        i=0
        for p in pts:
            s=e
            e=p
            if i%2==1:
                img = cv2.line(img,s,e,color,thickness)
            i+=1
    return img

def drawpoly(img,pts,color,thickness=1,style='dotted',):
    s=pts[0]
    e=pts[0]
    pts.append(pts.pop(0))
    for p in pts:
        s=e
        e=p
        drawline(img,s,e,color,thickness,style)

def drawrect(img,pt1,pt2,color,thickness=1,style='dotted'):
    pts = [pt1,(pt2[0],pt1[1]),pt2,(pt1[0],pt2[1])] 
    drawpoly(img,pts,color,thickness,style)
```

```python
def short_jaw(face_demo, landmarks, landmarks_avg):
    jaw_demo = face_demo.copy()
    lips_lowerline = landmarks[54:61] #+ mouth_diff
    for dot in lips_lowerline:
        position0 = (int(dot[0]), int(dot[1]))
        jaw_demo = cv2.circle(jaw_demo, position0, radius=1, color=(255, 0, 0), thickness=-1)
    jaw_line = landmarks[7:10]
    major_axis = int((jaw_line[2][0] - jaw_line[0][0])/3)
    jaw_center = np.average(jaw_line, axis=0)
    minor_axis = int(jaw_line[1][1] - jaw_center[1])
    jaw_position = (int(jaw_center[0]), int(jaw_center[1]-minor_axis))
    minor_axis *= 1.5
    jaw_demo = cv2.ellipse(jaw_demo, jaw_position, (int(major_axis), int(minor_axis)),0,0,360,(255,0,0))
    arrow_position_r = (landmarks[54] + landmarks[10])/2
    arrow_start_r = (int(arrow_position_r[0]-10), int(arrow_position_r[1]+10))
    arrow_end_r = (int(arrow_position_r[0]+10), int(arrow_position_r[1]-10))
    arrow_position_l = (landmarks[60] + landmarks[6])/2
    arrow_start_l = (int(arrow_position_l[0]+10), int(arrow_position_l[1]+10))
    arrow_end_l = (int(arrow_position_l[0]-10), int(arrow_position_l[1]-10))
    jaw_demo = cv2.arrowedLine(jaw_demo, arrow_start_r, arrow_end_r, color=(255,0,0), thickness=1)
    jaw_demo = cv2.arrowedLine(jaw_demo, arrow_start_l, arrow_end_l, color=(255,0,0), thickness=1)
    plt.imshow(jaw_demo)
    suggestion = '沿下唇线红色点的下缘打高光进行提亮，手法为45°斜向上。在下巴处打高光。注意保持晕染自然。唇妆可以选择偏裸色咬唇妆。'
    return jaw_demo, suggestion
jaw_demo, suggestion = short_jaw(face_demo, landmarks, landmarks_avg)
print(suggestion)
```

```python
import cv2
import mediapipe as mp
import time
import pdb
mp_drawing = mp.solutions.drawing_utils
mp_face_mesh = mp.solutions.face_mesh
count = 0
drawing_spec = mp_drawing.DrawingSpec(thickness=1, circle_radius=1)
cap = cv2.VideoCapture("Videos/hua_face.mp4")
with mp_face_mesh.FaceMesh(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as face_mesh:
    #pdb.set_trace()
    while cap.isOpened():
        success, image = cap.read()
        if not success:
            #print("Ignoring empty camera frame.")
          # If loading a video, use 'break' instead of 'continue'.
            break
        
        # Flip the image horizontally for a later selfie-view display, and convert
        # the BGR image to RGB.
        image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
        # To improve performance, optionally mark the image as not writeable to
        # pass by reference.
        image.flags.writeable = False
        results = face_mesh.process(image)
        time.sleep(0.02)
        # Draw the face mesh annotations on the image.
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        if results.multi_face_landmarks:
            #pdb.set_trace()
            for face_landmarks in results.multi_face_landmarks:
                mp_drawing.draw_landmarks(
                    image=image,
                    landmark_list=face_landmarks,
                    connections=mp_face_mesh.FACE_CONNECTIONS,
                    landmark_drawing_spec=drawing_spec,
                    connection_drawing_spec=drawing_spec)
        cv2.imwrite(f'Videos/hua_face_ex/frame{count}.jpg', image)
        count += 1
    #cv2.imshow('MediaPipe FaceMesh', image)
    #if cv2.waitKey(5) & 0xFF == 27:
    #    break
cap.release()
```
